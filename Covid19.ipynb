# Importando bibliotecas
Importando as bibliotecas que utilizaremos e definindo funções iniciais 

```
import pandas as pd
import numpy as np
import plotly.offline as py
import plotly.graph_objs as go

py.init_notebook_mode(connected=False)
```

```
# Devido o Plotly estar sendo utilizado no Google Collab, precisaremos definir
# a função abaixo e chamá-la sempre que quisermos exibir um gráfico

def configure_plotly_browser_state():
  import IPython
  display(IPython.core.display.HTML('''
        <script src="/static/components/requirejs/require.js"></script>
        <script>
          requirejs.config({
            paths: {
              base: '/static/base',
              plotly: 'https://cdn.plot.ly/plotly-1.5.1.min.js?noext',
            },
          });
        </script>
        '''))
```

# **Lendo DataSet**

```
#Dando permissão para acesso ao drive
from google.colab import drive
drive.mount('/content/drive')
```

```
dataset = pd.read_csv('/content/drive/My Drive/CovidAtualizado.csv', sep=";", encoding='latin1')
```

```
#Testando leitura do dataset
dataset.head()
```

# **Aqui teremos o crescimento dos casos novos por estado. Utilizaremos o gráfico de Dispersão Simples, para mostrar o crescimento gradual dos casos**

```
configure_plotly_browser_state()

trace = go.Scatter(x = dataset['estado'],
                   y = dataset['casosNovos'],
                   mode = 'markers')

data = [trace]

py.iplot(data)
```


# **Aqui teremos o número de casos totais por estados e em seguida por região. Utilizaremos o gráfico de barras para mostrar o total de números de infectados (com base no dataset) em cada lugar.**

```
configure_plotly_browser_state()
trace = go.Bar(x = dataset['estado'],
                   y = dataset['casosAcumulados'])
data = [trace]
py.iplot(data)
```

# **Aqui teremos o número de casos fatais totais por estado. Utilizaremos o gráfico de barras para mostrar o total de números de mortos (com base no dataset) em cada lugar.**

```
configure_plotly_browser_state()

trace = go.Bar(x = dataset['estado'],
                   y = dataset['obitosAcumulados'],
                   )

data = [trace]

py.iplot(data)
```

```
configure_plotly_browser_state()

trace = go.Bar(x = dataset['data'],
                   y = dataset['obitosAcumulados'],
                   )

data = [trace]

py.iplot(data)
```

# **Numero de casos de hospitalizações separadas por idade e sexo**

```
dataset2 = pd.read_csv('/content/drive/My Drive/Pasta1.csv', sep=";")
```

```
configure_plotly_browser_state()

trace1 = go.Scatter(x = dataset2['Idade'],
                    y = dataset2['Mulheres'],
                    mode = 'lines',
                    name = 'Mulheres',
                    line = {'color': 'red',
                            'dash': 'dash'})

trace2 = go.Scatter(x = dataset2['Idade'],
                    y = dataset2['Homens'],
                    mode = 'lines',
                    name = 'Homens',
                    line = {'color': 'green',
                            'dash': 'dot'})

data = [trace1, trace2]

py.iplot(data)
```
